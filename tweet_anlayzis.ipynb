# Text preprocessing

```python
#Preprocessing text

# FUNCTION FOR DELETE PUNCTUATION, TRANSFORM TEXT IN LOWERCASE, DELETE HYPER LINK,DELETE OTHER CARACTERE,
# TOKEN TEXT AND DELETE STOP WORDS
import string
import re
import nltk
#download stopwords from  nltk 
nltk.download('stopwords')

stopWord = nltk.corpus.stopwords.words("english")

df["tweet"] = df["tweet"].apply(lambda x: re.sub('http[s]?://\S+',' ', x))
df["tweet"] = df["tweet"].apply(lambda x: re.sub('[\W_]+', ' ', x))
df.head(20)
```

```python
def change_text(text):
    output = "".join([i for i  in text if i not in string.punctuation])
    return output

df["tweet"] = df["tweet"].apply(lambda x: change_text(x))
df.head(20)
```

```python
df["tweet"] = df["tweet"].apply(lambda x: x.lower())
df.head(20)
```

```python
df["tweet"] = df["tweet"].apply(lambda x: re.split("W+", x))
df.head(20)
```

```python
def delete_stopword(text, stopWord = stopWord):
    output = [i for i in text if i not in stopWord]
    return output
    
df["tweet"] = df["tweet"].apply(lambda x: delete_stopword(x))
df.head(20)
```

```python
#lemmatize text
from nltk.stem import WordNetLemmatizer

wordLemmatize = WordNetLemmatizer()

def lemmatizeWord(text, wordLemmatize = wordLemmatize):
    output = [wordLemmatize.lemmatize(word) for word in text]
    return output

df["tweet"] = df["tweet"].apply(lambda x: lemmatizeWord(x))
df.head(20)
```

```python
#Pour effectuer transformer les textes en vecteur il est important de faire en sorte qu'ils ne soit pas des listes
df["tweet"] = df["tweet"].apply(lambda x: " ".join(x))
df.head(20)
```

# Transform dataset on Vector

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
df["emotion"] = label.fit_transform(df["emotion"])

y = df["emotion"]
X = df["tweet"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

"""
vectorizer = CountVectorizer()
output_vectorizer = vectorizer.fit_transform(X)

frequency = TfidfTransformer()
X_other = frequency.fit_transform(output_vectorizer)

X_other.shape
"""
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
print(f" shape train: X= {X_train.shape}, y= {y_train.shape}  \n shape test: X= {X_test.shape}, y= {y_test.shape}")
```

# Modeling 

```python
#creation des modeles 
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.pipeline import Pipeline

def pipelineModel(model):
    pipeModel = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('TfTransformer', TfidfTransformer()),
    ('model_svc', model)])
    return pipeModel

#Pipeline for model SVC
svc = SVC()
modelSVC = pipelineModel(svc)

#Pipeline for model LinearSVC
linearSVC = LinearSVC()
modelLinearSVC = pipelineModel(linearSVC)

#Pipeline for model KNeighborsClassifier
KNeighborsclassifier = KNeighborsClassifier()
modelKnei = pipelineModel(KNeighborsclassifier)

#Pipeline for model RandomForestClassifier
RandomForestclassifier = RandomForestClassifier()
modelRandom = pipelineModel(RandomForestclassifier)

#Pipeline for model GradientBoostingClassifier
GradientBoosting = GradientBoostingClassifier()
modelGradient = pipelineModel(GradientBoosting)
```

```python
# fit modele

#modele SVC
modelSVC.fit(X_train, y_train)

#modele LinearSVC
modelLinearSVC.fit(X_train, y_train)

#modele KNeighborsClassifier
modelKnei.fit(X_train, y_train)

#modele randomforestclassifier
modelRandom.fit(X_train, y_train)

#modele gradientboostingclassifier
modelGradient.fit(X_train, y_train)
```

```python
#tester et entrainer les modeles
from sklearn.metrics import confusion_matrix, classification_report

#une fonction permettant d'effectuer tout ceci
def test_evaluate(y_test, y_pred):
    print(f"Classification report: \n {classification_report(y_test, y_pred)}")
    print(f"confusion matrix: \n {confusion_matrix(y_test, y_pred)}")
    
```

```python
#modele SVC

y_predSVC = modelSVC.predict(X_test)
test_evaluate(y_test, y_predSVC)
```

```python
#modele LinearSVC

y_predLinear = modelLinearSVC.predict(X_test)
test_evaluate(y_test, y_predLinear)
```

```python
#modele Kneighbors

y_predKneighbors = modelKnei.predict(X_test)
test_evaluate(y_test, y_predKneighbors)
```

```python
#modele RANDOM FOREST 

y_predRandom = modelRandom.predict(X_test)
test_evaluate(y_test, y_predRandom)
```

```python
#modele Gradient Boosting 

y_predGradient = modelGradient.predict(X_test)
test_evaluate(y_test, y_predGradient)
```
